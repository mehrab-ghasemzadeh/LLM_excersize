```python
import re
```

```python
textFilePath = "./text.txt"
with open (textFilePath,'r',encoding="utf-8") as file:
    text = file.read()
```

```python
text = text.strip()
```

```python
preprocessedVocab = re.split(r'([,.?_!"()\']|--|\s)',text)
len(preprocessedVocab)
```

```python
preprocessedVocab = sorted(list(set(preprocessedVocab)))
len(preprocessedVocab)
```

```python
for index,token in enumerate(preprocessedVocab):
    print(f"{index} - {token}")
```

```python
vocab = {token:index for index,token in enumerate(preprocessedVocab)}
len(vocab)
```

```python
inputText = "beneath their palm-trees; and Mrs. Gisburn's welcome was so genial"
```

```python
class tokenizer:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {index:token for token,index in vocab.items()}
    def encode(self, inp_text):
        preprocessed = re.split(r'([,.?_!"()\']|--|\s)', inp_text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        ids = [self.str_to_int[token] for token in preprocessed]
        return ids
    def decode(self, ids): 
        text = " ".join([self.int_to_str[i] for i in ids])
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text).replace("' ","'")
        return text
```

```python
t = tokenizer(vocab)
try:
    ids = t.encode(inputText)
    s = t.decode(ids)
    print(ids,s)
except Exception as e:
    print(f"ERROR - {e}")
```
